## 回归任务有监督模型综合对比分析
回归任务有监督模型综合对比分析需要同学们自行搜集回归任务的数据集（一个即可），使用学过的线性回归、决策树、神经网络、支持向量机四个模型（可以使用Scikit-learn或其他工具包）在该数据集上进行对比评测，要求：
1.	每个方法尝试不同参数，运用MSE,RMSE,MAE三个指标进行对比评测，把结果绘制成表格（四个模型四个表格）和可视化图（只可视化数值型参数即可，如SVM的C，决策树的深度等，每个模型每个参数一张图）
2.	运用第1条尝试得到的最优参数对四个模型进行横向对比，计算MSE,RMSE,MAE三个指标，把结果绘制成表格

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

#### 使用自带的sklearn.datasets.load_boston()  数据集

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

boston = load_boston()
data = boston.data
target = boston.target

trainX, testX, trainY, testY = train_test_split(data, target, test_size = 0.3, random_state = 32)

trainX.shape, trainY.shape, testX.shape, testY.shape
```

### 多元线性回归
MAE|MSE|RMSE
-|-|-
3.9973|34.5397|5.877

```python
from sklearn.linear_model import LinearRegression
linearModel = LinearRegression()
y_pred = cross_val_predict(linearModel, data, target, cv = 10)

MAE = round(mean_absolute_error(target, y_pred), 4)
MSE = round(mean_squared_error(target, y_pred), 4)
RMSE=round(mean_squared_error(target, y_pred)** 0.5,4)
print(MAE)
print(MSE)
print(RMSE)
```

### 神经网络MLP

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
trainX, testX, trainY, testY = train_test_split(data, target) 
# 初始化一个标准化器的实例
standard = StandardScaler()
# 对训练集进行标准化，它会计算训练集的均值和标准差保存起来
data = standard.fit_transform(data)



MAE = []
MSE = []
RMSE = []
for i in range(1,21):
    MLPmodel=MLPRegressor(hidden_layer_sizes=(i,),
                       activation='relu',
                       solver='adam',
                       learning_rate='adaptive',
                       max_iter=1000,
                       learning_rate_init=0.01,
                       alpha=0.01)
    y_pred = cross_val_predict(MLPmodel, data, target)
    MAE = np.append(MAE,round(mean_absolute_error(target, y_pred), 4))
    MSE = np.append(MSE,round(mean_squared_error(target, y_pred), 4))
    RMSE = np.append(RMSE,round(mean_squared_error(target, y_pred)** 0.5,4))
    
print(MAE)
print(MSE)
print(RMSE)

x = np.arange(1,21,1)
plt.plot(x,MAE,'r',label='MAE')
plt.plot(x,MSE,'g',label='MSE')
plt.plot(x,RMSE,'b',label='RMSE')
plt.xlabel('hidden_layer_sizes = (x,)')
plt.ylabel('性能')
plt.legend()
```
