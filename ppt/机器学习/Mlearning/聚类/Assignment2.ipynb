# 第二题：使用sklearn的KMeans完成聚类

实验内容：
1. 使用sklearn的Kmeans完成两个数据集的聚类任务
2. 计算外部指标FMI和NMI
3. 对聚类结果可视化

# 1. 导入模块

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# 2. 生成数据集

```python
from sklearn import datasets
X1, y1 = datasets.make_circles(n_samples = 1500, factor = 0.5, noise = 0.05, random_state = 32)
X2, y2 = datasets.make_moons(n_samples = 1500, noise = 0.05, random_state = 32)
```

```python
colors = np.array(['#377eb8', '#ff7f00', '#4daf4a', '#f781bf', '#a65628', '#984ea3', '#999999', '#e41a1c', '#dede00'])
```

```python
plt.title('dataset1')
plt.scatter(X1[:, 0], X1[:, 1], s = 10, c = colors[y1])
```

```python
plt.title('dataset2')
plt.scatter(X2[:, 0], X2[:, 1], s = 10, c = colors[y2])
```

# 3. 导入模型

我们导入KMeans模型

```python
from sklearn.cluster import KMeans
```

# 4. 聚类

```python
model_1 = KMeans(n_clusters=2)
model_1.fit(X1)
```

```python
model_2 = KMeans(n_clusters=2)
model_2.fit(X2)
```

使用predict方法预测结果

```python
model_1.predict(X1)
```

```python
model_2.predict(X2)
```

# 5. 聚类结果可视化

```python
plt.title('dataset1')
plt.scatter(X1[:, 0], X1[:, 1], s = 10, color=colors[model_1.predict(X1)])
```

```python
plt.title('dataset2')
plt.scatter(X2[:, 0], X2[:, 1], s = 10, color=colors[model_2.predict(X2)])
```
